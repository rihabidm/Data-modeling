# 3.1. Détéction des valeurs aberrantes

```python
plt.figure(figsize=(15, 5))
plt.subplot(1,4,1)
plt.boxplot(titanic['age'].dropna(), sym='g+')
plt.title('Boxplot of Age')
plt.subplot(1,4,2)
plt.boxplot(titanic['parch'].dropna(), sym='g+')
plt.title('Boxplot of parch')
plt.subplot(1,4,3)
plt.boxplot(titanic['sibsp'].dropna(), sym='g+')
plt.title('Boxplot of sibsp')
plt.subplot(1,4,4)
plt.boxplot(titanic['pclass'].dropna(), sym='g+')
plt.title('Boxplot of Pclass')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

Q1 = np.percentile(titanic['age'].dropna(), 25)
Q3 = np.percentile(titanic['age'].dropna(), 75)
IQR = Q3 - Q1

outliers = titanic[(titanic['age'] < Q1 - 1.5 * IQR) | (titanic['age'] > Q3 + 1.5 * IQR)]
print("Outliers:\n", outliers[['age']])
```

## 3.2.Traitement des valeurs aberrantes

### 3.2.1 Remplacer les valeurs aberrantes pour les deux extrémités 

```python
# Replacing outliers with bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data_imputed['age'] = np.where(data_imputed['age'] < lower_bound, lower_bound, data_imputed['age'])
data_imputed['age'] = np.where(data_imputed['age'] > upper_bound, upper_bound, data_imputed['age'])
```

```python
numerical_data = data_imputed.select_dtypes(include=['number'])
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
